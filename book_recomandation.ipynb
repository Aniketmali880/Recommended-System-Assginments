```python
import pandas as pd
import numpy as np
```

```python
## loading the dataset
book_1=pd.read_csv("Book.csv",encoding='windows-1252')
```

```python
book_1.head()
```

```python
len(book_1.book.unique())
```

```python
## Making the pivot table
user_book_df = pd.pivot_table(book_1,index='userId',
                                 columns='book',
                                 values='rating').reset_index(drop=False)
```

```python
user_book_df
```

```python
user_book_df.index = book_1.userId.unique()
```

```python
user_book_df
```

```python
## Filling NA with 0
user_book_df.fillna(0, inplace=True)
```

```python
user_book_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book_1.userId.unique()
user_sim_df.columns = book_1.userId.unique()
```

```python
user_sim_df.iloc[0:10, 0:10]
```
